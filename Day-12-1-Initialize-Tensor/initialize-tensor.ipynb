```python
import numpy as np
import torch
import tensorflow as tf
```

### 1. Zeros() Method

```python
# Numpy syntax
arr_zeros = np.zeros((3,4))
print(arr_zeros)

# Pytorch Syntax
tensor_zeros_pt = torch.zeros((3,4))
print(tensor_zeros_pt)

# Tensorflow syntax
tensor_zeros_tf = tf.zeros((3,4))
print(tensor_zeros_tf)
```

### 2. Ones() Method

```python
# Numpy syntax
arr_ones = np.ones((3,4))
print(arr_ones)

# Pytorch Syntax
tensor_ones_pt = torch.ones((3,4))
print(tensor_ones_pt)

# Tensorflow syntax
tensor_ones_tf = tf.ones((3,4))
print(tensor_ones_tf)
```

### 3. Full / Fill

```python
# Numpy syntax
arr_full = np.full((3,4), 5)
print(arr_full)

# Pytorch Syntax
tensor_full_pt = torch.full((3,4), 5)
print(tensor_full_pt)

# Tensorflow syntax
tensor_fill_tf = tf.fill((3,4), 5)
print(tensor_fill_tf)
```
