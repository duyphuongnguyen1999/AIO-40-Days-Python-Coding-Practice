Ví dụ này nhằm mục đích giúp ae hiểu rằng các thư viện này có thể chuyển đổi lẫn nhau trong quá trình tính toán nên mọi người cứ thoải mái dùng 3 thư viện này, đôi khi có thể kết hợp chúng khi cần thiết.

Ae có thể chạy chương trình này và tinh chỉnh thêm một số trường hợp khác sẽ nhận thấy rằng kiểu dữ liệu của kết quả đầu ra sẽ theo thứ tự ưu tiên Tensorflow >Pytorch > Numpy. Và cách trong ví dụ hoạt động với một số phép tính sử dụng toán tử cơ bản như +, -, *, /...trong trường hợp bên dưới còn có thể thực hiện nhân ma trận giữa tensor pytorch và array numpy với toán tử @.

Lưu ý: Cách trên không hoạt động khi sử dụng các hàm hoặc phương thức riêng của mỗi thư viện. Để tránh nhầm lẫn gây ra lỗi thì ae vẫn nên chuyển đổi các cấu trúc dữ liệu của các thư viện khác nhau về chung một thư viện để dễ dàng xử lý trong các trường hợp phức tạp.... Và chắc còn nhiều thứ nữa em chưa phát hiện ra, ae thử các trường hợp rồi comment cho mọi người cùng biết nhé ạ!

```python
import numpy as np
import torch
import tensorflow as tf

array_np = np.array([1, 2, 3])
tensor_pt = torch.tensor([4, 5, 6])
tensor_tf = tf.constant([7, 8, 9])

result = tensor_tf - (tensor_pt @ array_np) * array_np
print(result)
```
