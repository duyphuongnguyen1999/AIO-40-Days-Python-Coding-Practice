```python
import numpy as np
import torch
import tensorflow as tf
```

### Initialize an Array, Tensor

```python
list_1D = [x for x in range(1, 10)]
# Numpy syntax
arr_1D = np.array(list_1D)
print("Numpy 1D-Array:", arr_1D, type(arr_1D))
# Pytorch syntax
tensor_1D_pt = torch.tensor(list_1D)
print("Pytorch 1D-Tensor:", tensor_1D_pt, type(tensor_1D_pt))
# Tensorflow syntax
# tensor_1D_tf = tf.constant(list_1D)
tensor_1D_tf = tf.convert_to_tensor(list_1D)
print("Tensorflow 1D-Tensor:", tensor_1D_tf, type(tensor_1D_tf))
```

```python
list_2D = [[1, 2, 3],
           [4, 5, 6],
           [7, 8, 0]]
# Numpy syntax
arr_2D = np.array(list_2D)
print("Numpy 2D-Array:", arr_2D)
print("Shape: ", arr_2D.shape)
print("Data type: ", arr_2D.dtype)
print("Array (Tensor) Type: ", type(arr_2D))

# Pytorch syntax
tensor_2D_pt = torch.tensor(list_2D)
print("Pytorch 2D-Tensor:", tensor_2D_pt)
print("Pytorch 2D-Tensor:", tensor_2D_pt)
print("Shape: ", tensor_2D_pt.shape)
print("Data type: ", tensor_2D_pt.dtype)
print("Array (Tensor) Type: ", type(tensor_2D_pt))

# Tensorflow syntax
# tensor_2D_tf = tf.constant(list_2D)
tensor_2D_tf = tf.convert_to_tensor(list_2D)
print("Tensorflow 2D-Tensor:", tensor_2D_tf)
print("Tensorflow 2D-Tensor:", tensor_2D_tf)
print("Shape: ", tensor_2D_tf.shape)
print("Data type: ", tensor_2D_tf.dtype)
print("Array (Tensor) Type: ", type(tensor_2D_tf))
```
